```python
def compare_performance(metrics, best_metrics):
    return metrics[0] >= best_metrics[0] and metrics[1] >= best_metrics[1]
```

```python
val_pred = val.copy()
```

```python
results = []
for lower_bound in range(60,121, 1):
    for upper_bound in range(120, 59, -1):
        if lower_bound >= upper_bound:
            break
        print('processing ' + str(lower_bound) + ' ' + str(upper_bound))
        bidprices = np.random.randint(lower_bound, upper_bound+1, size=val_pred['bidprice'].shape)
        val_pred['bidprice'] = bidprices
        budget = 6250
        suc_bids = val_pred.query('bidprice >= payprice and bidprice >= slotprice')
        cost = 0
        clicks = 0
        imps = 0
        for index, row in suc_bids.iterrows():
            if cost <= budget:
                cost += row['payprice'] / 1000
                clicks += row['click']
                imps += 1
            
        eCPC = cost / clicks if clicks > 0 else float('inf')
        metrics = [clicks, clicks / imps * 100, cost, cost / imps * 1000, eCPC]
        results.append([lower_bound, upper_bound, clicks, clicks / imps * 100, cost, cost / imps * 1000, eCPC])
results = pd.DataFrame(results)
results = results.rename(columns={0: "lower_bound", 1: "upper_bound", 2: "clicks",
    3: "CTR", 4: "cost", 5: "avg CPM", 6: "eCPC"})
results.to_csv(os.path.join('random_bid.csv'), index=False)
```
